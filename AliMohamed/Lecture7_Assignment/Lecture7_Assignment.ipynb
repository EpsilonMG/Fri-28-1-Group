## Task 1 

```python
import requests
import csv
import pandas as pd
```

```python
#We can change the base amount of USD with 2 different ways
#First, we change the amount in the API url link
response1 = requests.request('GET','https://api.frankfurter.app/latest?amount=10&from=USD&to=EUR')
data1 = response1.json()
data1
```

```python
print(f"WELCOME, HERE YOU CAN CONVERT USD TO EUR. DUE TO ({data1['date']})")
amount = data1['amount']
rate = data1['rates']['EUR']
print(f'FOR {amount} USD = {rate} EUR')
```

```python
#Second, by calculating the amount from the data itself
response2 = requests.request('GET','https://api.frankfurter.app/latest?amount=1&from=USD&to=EUR')
data2 = response2.json()
data2
```

```python
print(f"WELCOME, HERE YOU CAN CONVERT USD TO EUR. DUE TO ({data2['date']})")
rate = data2['rates']['EUR']
try:
    amount = int(input('ENTER AN AMOUNT : '))
    result = amount * rate
    print(f'USD AMOUNT TO EUR = {result}')
except Exception as e:
    print('A TYPE ERROR HAS OCCURED. PLEASE TRY AGAIN')
```

## Task 2

```python
response3 = requests.request('GET','https://www.metaweather.com/api/location/1521894/')
data3 = response3.json()
data3
```

```python
with open('weather.csv','w',newline='') as f:
    writer = csv.DictWriter(f,fieldnames=['Weather temperature', 'Humidity', 'Visibility', 'Air pressure', 'Wind speed', 'Date time'])
    writer.writeheader()
    for weather in data3['consolidated_weather']:
        writer.writerow({'Weather temperature':weather['the_temp'], 'Humidity':weather['humidity'], 'Visibility':weather['visibility'], 'Air pressure':weather['air_pressure'], 'Wind speed':weather['wind_speed'], 'Date time':weather['applicable_date']})
        
```

```python
dataframe = pd.read_csv('weather.csv')
dataframe
```

## Task 3

```python
response4 = requests.request('GET','https://jsonplaceholder.typicode.com/users')
data4 = response4.json()
data4
```

```python
with open('users.csv','w',newline='') as f:
    writer = csv.DictWriter(f,fieldnames=['Name', 'Username', 'E-mail', 'Street', 'Suite', 'City', 'Zip code', 'geo_lat', 'geo_long'])
    writer.writeheader()
    for user in data4:
        writer.writerow({'Name':user['name'], 'Username':user['username'], 'E-mail':user['email'], 'Street':user['address']['street'], 'Suite':user['address']['suite'], 'City':user['address']['city'], 'Zip code':user['address']['zipcode'], 'geo_lat':user['address']['geo']['lat'], 'geo_long':user['address']['geo']['lng']})
        
```

```python
dataframe = pd.read_csv('users.csv')
dataframe
```
