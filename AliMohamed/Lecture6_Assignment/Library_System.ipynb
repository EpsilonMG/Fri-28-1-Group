```python
class libaray:
    def __init__(self,l):
        self.list = l
    def add_book(self,bookName):
        self.list.append(bookName)
        print(f'New book \"{bookName}\" is added.')
    def give_book(self,bookName):
        if bookName in self.list:
            self.list.remove(bookName)
            print(f'\"{bookName}\" book is taken.')
            return True
        else:
            print(f'\"{bookName}\" isn\'t available or incorrect.')
            return False
    def showBooks(self):
        print(f'Available Books: {self.list}')
```

```python
class Person:
    def __init__(self,fname,lname):
        self.fname = fname
        self.lname = lname
    def borrow(self,lib):
        bookName = input('Enter Book Name: ')
        state = lib.give_book(bookName)
        if state:
            print(f'Successful borrow from {self.fname} {self.lname}, Have Fun!')
        else:
            print(f'Failed to borrow {bookName}')
    def deliever(self,lib):
        lib.add_book(input('Enter Book Name: '))
        print(f'Successful deliever from {self.fname} {self.lname}, Thank you!')
```

```python
class Admin(Person):
    def __init__(self,fname,lname,user,password):
        self.fname = fname
        self.lname = lname
        self.user = user
        self.role = 'Admin'
        self.ps = password
```

```python
class Student(Person):
    def __init__(self,fname,lname,user,password):
        self.fname = fname
        self.lname = lname
        self.user = user
        self.role = 'Student'
        self.ps = password
```
