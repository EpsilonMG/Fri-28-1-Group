```python
class library:    
    def __init__(self, books):
        self.lst_books = books
      
    def list_books(self):
        print("Available list of books \n", ' '.join(self.lst_books))
        
    def add_book(self, book):
        if book not in self.lst_books:
            self.lst_books.append(book)
            print(book," Added")
        else:
            print(book," is already in liberary")
            
    def remove_book(self, book):
        if book in self.lst_books:
            self.lst_books.remove(book)
            print(book," Lended")
        else:
            print(book,"is not available")

        
class student:
    def return_book(self):
        book = input("Please enter book name: ")
        return book
    
    def lend_book(self):
        book = input("Please enter book name: ")
        return book
```

```python
lib=library(['DataScience', 'BigData', 'MachineLearning', 'DataBase', 'DeepLearning', 'F'])
std=student()

while True:
    try:
        selected_option = int(input("""Hello in liberary system, check the below options:
        1) List Available Books
        2) Lend Book
        3) Return Book
        4) Exit
        """))
        if selected_option == 1:
            lib.list_books()
        elif selected_option == 2:
            book_name = std.lend_book()
            lib.remove_book(book_name)
        elif selected_option == 3:
            book_name = std.return_book()
            lib.add_book(book_name)
        elif selected_option == 4:
            print("System Exit")
            break
        else:
            print("Invalid Input")
    except:
        print("Invaild Input Type")
```
