```python
import os
from pathlib import Path

class library:   
    def __init__(self):
        self.file_name = 'data.txt'
        main_book_lst = ['Data Science\n','Machine Leaning\n',
                         'Deep Learning\n','Database\n','Big Data\n',
                         'NLP\n','RNN\n','CNN\n']
        is_exist = os.path.exists(str(Path().absolute())+'/'+self.file_name)
        if not is_exist:
            with open(self.file_name, 'w+') as file:
                file.writelines(main_book_lst)

    def list_books(self):
        with open(self.file_name, 'r+') as file:
            lines=file.read()
            print("Available books:\n",''.join(lines))
        
    def add_book(self, book):
        with open(self.file_name, 'r+') as file:
            content=file.read()
            if book not in content.split('\n'):
                file.write(book+'\n')
                print(book," is Added")
            else:
                print(book," is already in library")
            
    def remove_book(self, book):
        with open(self.file_name, 'r+') as file:
            lines = file.read().split('\n')
            book_index = [index for index, line in enumerate(lines) if line==book]
            if len(book_index) > 0:
                lines.remove(lines[book_index[-1]])
                file.seek(0)
                file.truncate(0)
                lines = '\n'.join(lines)
                file.writelines(lines)
                print(book," is Removed")
            else:
                print(book,"is not available")
            
        
class student:
    def return_book(self):
        book = input("Please enter book name: ")
        return book
    
    def borrow_book(self):
        book = input("Please enter book name: ")
        return book
```

```python
lib=library()
std=student()

while True:
    try:
        selected_option = int(input("""Hello in liberary system, check the below options:
        1) List Available Books
        2) Borrow Book
        3) Return Book
        4) Exit
        """))
        if selected_option == 1:
            lib.list_books()
        elif selected_option == 2:
            book_name = std.borrow_book()
            lib.remove_book(book_name)
        elif selected_option == 3:
            book_name = std.return_book()
            lib.add_book(book_name)
        elif selected_option == 4:
            print("System Exit")
            break
        else:
            print("Invalid Input")
    except Exception as e:
        print(e)
```
